### Feature Engineering

```python
feat_eng = ['A1_Score', 'A2_Score', 'A3_Score', 'A4_Score', 'A5_Score', 'A6_Score', 'A7_Score', 'A8_Score', 'A9_Score', 'age', 'result']

def add_features(data):
    for col in feat_eng:
        feat = str(col) + '_mean'
        data[feat] = data[col].mean()
        feat = str(col) + '_sum'
        data[feat] = data[col].sum()
        feat = str(col) + '_std'
        data[feat] = data[col].std()
        
```

## Modeling

```python
del data['ID']
X = data.copy()
#X = data[features]
y = X.pop('Class/ASD')
#y = data['Class/ASD']

X.shape
```

```python
!pip install flaml
```

```python
from flaml import AutoML
automl = AutoML()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
automl.fit(X, y, task="regression", metric='rmse', time_budget=1200) # 20 min
```

```python
print('Best ML leaner:', automl.best_estimator)
print('Best hyperparmeter config:', automl.best_config)
print('Best log_loss on validation data: {0:.4g}'.format(automl.best_loss))
print('Training duration of best run: {0:.4g} s'.format(automl.best_config_train_time))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
pred = automl.predict(X_test)

mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)

print("MAE: %.7f" % mae)
print("MSE: %.7f" % mse)
```
